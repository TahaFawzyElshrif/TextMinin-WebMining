```python
#getting into one function

from bs4 import BeautifulSoup
import urllib.request
import re


def scarpeFile(linkWeb,name,pathFolders):#pathFolders with dataTXT (specify folders ended with /)
    url=urllib.request.urlopen(linkWeb).read()
    soup = BeautifulSoup(url, 'html.parser')
    for tag in soup.find_all(['h1', 'h2', 'h3', 'a','br','button','ul']):#h4,h5 ,h6 may be important
        tag.extract()

    #delete parts
    #search
    div = soup.find("div", {"class": "vector-header-end"})
    if div is not None:
      div.remove
    #right side
    div = soup.find_all("nav", {"id": "mw-panel-toc"})
    if div is not None:
      div.remove
    #menu
    div = soup.find_all("nav", {"id": "vector-main-menu-landmark"})
    if div is not None:
      div.remove

    #----------------->there is a problem :right part,may have important info is discarded
    div = soup.find_all("table", {"class":"infobox"})
    if div is not None:
      div.remove



    text=soup.get_text()

    #delete spaces ,tabs
    text = re.sub("\n", "", text)
    text = re.sub("\t", "", text)

    try:#to work you must create dataTXT
        f = open((pathFolders+name+".txt"), "w")
    except FileNotFoundError:
        f = open((pathFolders+name+".txt"), "x")#x - Create - will create a file, returns an error if the file exist
        f = open((pathFolders+name+".txt"), "w")#w - Write - will create a file if the specified file does not existf.write(text)
    f.write(text)
    f.close()
```

```python
#creating corpus
import os
corpus=dict({})
for path, name, files in os.walk(path_drive):#path=path_drive ,files=list of names of "files" in path
  for filename in files:
    with open(os.path.join(path,filename),"r") as f:#r : opened files here for reading only
      corpus[filename]=f.read()
print(corpus['Nitrogen(N)--p.txt'])#for test
```

```python
#preparing documents
#updating corpus to be words instead of total text ,and noromalized
#i didn't make another variable to optimize space
import re
import spacy

for fileKey in corpus:
  #lower case
  corpus[fileKey]=corpus[fileKey].lower()

  #punctions
  punctuations = r"[!@#$%^&*()_+{}|:\"<>?,./]"
  corpus[fileKey] = re.sub(punctuations, "", corpus[fileKey])

  #stopwords remove
  nlp = spacy.load("en_core_web_sm")
  doc = nlp(corpus[fileKey])
  #corpus[fileKey] = [token.text for token in doc if not token.is_stop]
  #corpus[fileKey] = [nlp(token).lemma_ for  token in corpus[fileKey]]
  words_normalized={}
  for token in doc:
    if not token.is_stop:
      token_nonStop=token.text  #stopwords remove
      token_lemma=nlp(token_nonStop)[-1].lemma_#nlp here on one word so last element in list is the word  #stemming and lemmatizition
      if token not in [" ", ]:
        words_normalized.append(token_lemma)

  corpus[fileKey]=words_normalized


  #Dictionary based like (car,train)->automobile  not suitable in this project due to chemical
  #corpus[fileKey]=re.split("\s+",corpus[fileKey]) this problem is solved above when removing stopwords , spacy is sophisticated  way so this line isn't important
#test
corpus['Cerium(Ce)--f.txt']
#don't run this cell 2 times consecutive to do so reassign corpus by last cell
```

```python
import json
import os
path_json="/content/drive/My Drive/projects/TextMining/Files/"#last path
corpus_string=""
with open(os.path.join(path_json,"corpus.json"),"r") as f:
  corpus_string=str(f.read())
corpus = json.loads(corpus_string)
corpus
```

```python
#Term frequency table
import pandas as pd

listOfALLWords=[]
for corpusListWords in corpus :
  listOfALLWords.extend(corpus[corpusListWords])

TF_DF=pd.DataFrame(len(listOfALLWords)*[0],index=listOfALLWords)
for doc in corpus:
  TF_DF[doc]=len(listOfALLWords)*[0]

#delete first non important column
TF_DF = TF_DF.iloc[:, 1:]



TF_DF
```

```python
#filling TF_DF
for doc in corpus:
  for word in listOfALLWords:
    TF_DF.loc[word,doc]=corpus[doc].count(word)/len(corpus[doc])

```

```python
#IDF table
N=len(corpus)# number of documents
IDF_dict=dict({})
#listOfALLWords used from above
for word in listOfALLWords:
  IDF_dict[word]=0
for word in listOfALLWords:
  for doc in corpus:
    if word in corpus[doc]:
      IDF_dict[word]=IDF_dict[word]+1
    IDF_dict[word]=IDF_dict[word]/N
IDF_dict
```

```python
#last table to dataframe
words=list(IDF_dict.keys())
counts=list(IDF_dict.values())
IDF_DF=pd.DataFrame(counts,index=words)
print(IDF_DF)
```

```python
#TF-IDF
TF_IDF_DF=pd.DataFrame(len(words)*[0],index=words)
for doc in corpus :
  TF_IDF_DF[doc]=len(words)*[0]

#delete first non important column
TF_IDF_DF = TF_IDF_DF.iloc[:, 1:]

#filling data of TF_IDF_DF
for doc in corpus:
  for word in words:
    TF_IDF_DF[doc][word]=TF_DF[doc][word]*pd.to_numeric(IDF_DF.loc[word])
TF_IDF_DF
```
